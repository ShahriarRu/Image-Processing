```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = 'images/cat.jpg'
```

```python
img = plt.imread(img_path)
```

```python
img.shape
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
original_hist = plt.hist(gray.ravel(),256,[0,256])
```

```python
#original_hist = cv.calcHist([gray],[0],None,[256],[0,256])
```

```python
equalized = cv2.equalizeHist(gray)
```

```python
equalized_hist = plt.hist(equalized.ravel(),256,[0,256])
```

```python
plt.subplots(figsize=(20,20))
plt.subplot(2,2,1)
plt.title("A. Original Image")
plt.imshow(img,cmap='gray')

plt.subplot(2,2,2)
plt.title("B. Original Histogram")
original_hist = plt.hist(gray.ravel(),256,[0,256])

plt.subplot(2,2,3)
plt.title("C. Equalized Image")
plt.imshow(equalized,cmap='gray')

plt.subplot(2,2,4)
plt.title("D. Equalized Histogram")
equalized_hist = plt.hist(equalized.ravel(),256,[0,256])
```
