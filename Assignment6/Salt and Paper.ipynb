# Importing Libraries

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import random
```

# Loading Image

```python
img_path = './images/flower.jpg'
img_path
```

```python
bgr_img = cv2.imread(img_path, -1)
rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
gray_img = cv2.imread(img_path, 0)
rgb_img.shape
gray_img.shape
```

# Declaring Functions

```python
def salt_paper_noise(img, thres):
    
    w,h = img.shape    
    output = img.copy()
    
    for i in range(w):
        for j in range(h):
            rand = random.random()
            if rand < thres:
                output[i,j] = 0
            #elif rand > 1-thres:
                #output[i,j] = 0
    return output
```

# Noisy Image

```python
noisy_img = salt_paper_noise(gray_img, 0.01)
```

# gaussian blur

```python
average_blur = cv2.blur(noisy_img,(5,5))
```

```python
gauss_blur = cv2.GaussianBlur(noisy_img,(5,5),0)
```

```python
median_blur = cv2.medianBlur(noisy_img,5)
```

```python
plt.figure(figsize=(30,25))
plt.subplot(3,2,1)
plt.title('RGB')
plt.imshow(rgb_img, cmap='gray')

plt.subplot(3,2,2)
plt.title('GRAY')
plt.imshow(gray_img, cmap='gray')

plt.subplot(3,2,3)
plt.title('NOISY')
plt.imshow(noisy_img, cmap='gray')

plt.subplot(3,2,4)
plt.title('Average Blur')
plt.imshow(average_blur, cmap='gray')

plt.subplot(3,2,5)
plt.title('Gauss Blur')
plt.imshow(gauss_blur, cmap='gray')

plt.subplot(3,2,6)
plt.title('Madian Blurr')
plt.imshow(median_blur, cmap='gray')

plt.savefig('nise_blur.png')
```
